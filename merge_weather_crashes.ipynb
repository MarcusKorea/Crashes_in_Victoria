```python
import pandas as pd
```

```python
# read melb data
melb_crash_path = "cleaned_data/melbourne_data.csv"
melb_weather_path = "cleaned_data/melbourne_weather.csv"
melb_crash = pd.read_csv(melb_crash_path)
melb_weather = pd.read_csv(melb_weather_path)

# read hume data
hume_crash_path = "cleaned_data/hume_data.csv"
hume_weather_path = "cleaned_data/hume_weather.csv"
hume_crash = pd.read_csv(hume_crash_path)
hume_weather = pd.read_csv(hume_weather_path)
```

```python
hume_crash.head()
hume_crash = hume_crash.rename(columns = {'ACCIDENT_DATE': "Date"})
```

```python
# rename accident_date to date
melb_crash = melb_crash.rename(columns = {'ACCIDENT_DATE': "Date"})
```

```python
# merge melb
merged_melb = melb_crash.merge(melb_weather, how ='inner', on = 'Date')

# merge hume crash and hume weather
merged_hume = hume_crash.merge(hume_weather, how ='inner', on = 'Date')
```

```python
merged_melb = merged_melb.drop(['OBJECTID','Unnamed: 0'],axis =1)
merged_hume = merged_hume.drop(['OBJECTID','Unnamed: 0'],axis = 1)
```

```python
merged_melb.to_csv("cleaned_data/merged_melb_data.csv", index = False)
merged_hume.to_csv("cleaned_data/merged_hume_data.csv", index = False)
```
